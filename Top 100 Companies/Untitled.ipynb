```python
def BeautifulSoup_data(input_data_content, par):
    soup = BeautifulSoup(input_data_content, par)
    return soup   
```

```python
def establish_connection(link):
    data=  fetch_data(link)
    print(data.status_code)
    while(data.status_code != 200):
        data = fetch_data(link)
        print('data not connected to source')
        if(data.status_code == 200):
            print('welcome data connect successfully')
            break
    return data
```

# list of data

```python
list_of_company_name= []
list_of_location=[]
list_of_package=[]
list_of_websites=[]
list_of_focus=[]
```

```python
counter = 0


for company_names in suop.find_all('div',{'class':"main-content has-sidebar"}):
    

    
    for i in company_names.find_all("p"):
        for j in i.find_all("strong", text="Location"):
            list_of_location.append(j.next_sibling)
            
        for j in i.find_all("strong", text="Focus Area"):
            list_of_focus.append(j.next_sibling)
            
        for j in i.find_all("strong", text="Average salary for data scientist:"):
            list_of_package.append(j.next_sibling)
            
        for j in i.find_all("strong", text="Average salary for data scientist"):
            list_of_package.append(j.next_sibling)
            
            
#             print("***************************************************************************************")            
        for j in i.find_all("strong", text="Website:"):
            list_of_websites.append(i.text)
            
        for j in i.find_all("strong", text="Website"):
            list_of_websites.append(i.text)
            
    for company_name in company_names.find_all('h4'):
        list_of_company_name.append(company_name.text)
```
