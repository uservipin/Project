```python
def establish_connection(link):
    data=  fetch_data(link)
    data.status_code
    while(data.status_code != 200):
        data = fetch_data(link)
        print('data not connected to source')
        if(data.status_code == 200):
            print('welcome data connect successfully')
            break
    return data
```

```python
def fetch_data(link):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
#     url= "https://wwlinkw.sec.gov/Archives/edgar/data/3662/0000950170-98-000413.txt"
    data = requests.get(link,headers)
    sleep(5)
    return data
```

# List to add values 

 Title of the product

• Ratings

• Number of reviews

• Price of the product

• Free delivery or not

```python
list_of_Title= list()
list_of_price= list()
list_free_delivery_or_not= list()


list_of_Ratings_and_reviews =list()
list_of_reviews =list()


```

```python
for  div in soup.find_all('div',{'id':"search"}):
    for i  in div.find_all('div', {'class':"s-matching-dir sg-col-16-of-20 sg-col sg-col-8-of-12 sg-col-12-of-16"}):
        for j in i.find_all('div', {'class':'a-section a-spacing-none'}):
#             print(j.prettify())
            
#     Tilte 
            for k in j.find_all('div',{'class':"a-section a-spacing-none a-spacing-top-small"}): 
                for m in k.find_all('h2', {'class':"a-size-mini a-spacing-none a-color-base s-line-clamp-2"}):
                    print(m.text)
                    list_of_Title.append(m.text)
                    
# Price 
                for l in k.find_all('div', {'class':'a-row a-size-base a-color-base'}):
                    for m in l.find_all('span', {'class': "a-price" ,'data-a-color' :"price"}):
                        for n in m.find_all('span', {'class' :"a-offscreen"}):
                            print(n.text)
                            list_of_price.append(n.text)
# Free delivery 
                for k in j.find_all('div',{'class' :"a-section a-spacing-none a-spacing-top-micro"}):
                    for m in k.find_all('span',{'aria-label':"FREE Delivery by Amazon"}):
                        print(m.text)
                        list_free_delivery_or_not.append(m.text)
                            
                            
# ratings and review     
                for k in j.find_all('div',{'class' :"a-section a-spacing-none a-spacing-top-micro"}):
                    for l in k.find_all('div',{'class': "a-row a-size-small"}):
                        for m in l.find_all('span'):
                            print(m.text)
                            list_of_Ratings_and_reviews.append(m.text)
                
```

# Cereate data frame of scraped data

#### Data is

####  Title of the product

#### Ratings

####  Number of reviews

####  Price of the product

#### Free delivery or not

### in csv file 

```python
dict_of_data_set = {
    'Product_Title':list_of_Title,
    'Price_of_product': list_of_price,
    'No_of_reviews': list_of_Reviews,
    'Rating':list_of_Ratings,
    'Free_delivery' : list_free_delivery_or_not
}
```

```python
final_data_set= pd.DataFrame.from_dict(dict_of_data_set, orient='index')


final_data_set= final_data_set.transpose()
```
