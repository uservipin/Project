```python
from datetime import datetime
from bs4 import BeautifulSoup as bs 
import requests
from time import sleep
import re
import pandas as pd
import matplotlib.pyplot as plt

```

```python

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
url = 'https://agmarknet.gov.in/SearchCmmMkt.aspx?Tx_Commodity=24&Tx_State=UP&Tx_District=1&Tx_Market=2551&DateFrom=06-Jul-2021&DateTo=16-Sep-2021&Fr_Date=06-Jul-2021&To_Date=16-Sep-2021&Tx_Trend=0&Tx_CommodityHead=Potato&Tx_StateHead=Uttar+Pradesh&Tx_DistrictHead=Agra&Tx_MarketHead=Achnera'
```

```python
resp= requests.get(url,headers)
resp.status_code
```

```python
soup = bs(resp.content,'html.parser')
# soup
print(soup.prettify())
```

```python
soup.title
```

```python
table = soup.find('table', class_="tableagmark_new")
```

```python
headers= []

for i in table.find_all('th'):
    title = i.text.strip()
    headers.append(title)
```

```python
headers
```

```python
list_of_2d_data= []
for row in table.find_all('tr')[1:]:
    list_of_data= []
#     print(row)
    for col in row.find_all('td'):
        list_of_data.append(col.text.strip())
    list_of_2d_data.append(list_of_data)
#     length = len(df)
#     df.loc[length]= row_data

# print(len(data))
```

```python
# print(list_of_2d_data)
```

```python
df= pd.DataFrame(list_of_2d_data, columns =headers)
```

```python
df
```

# Major Market 

```python
df['Modal Price (Rs./Quintal)'].dropna()
df['Modal Price (Rs./Quintal)'] = df['Modal Price (Rs./Quintal)'].astype(float)
```
