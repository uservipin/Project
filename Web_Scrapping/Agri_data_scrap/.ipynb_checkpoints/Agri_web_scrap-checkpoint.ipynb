```python
def get_soup(url):
    
    resp= requests.get(url,headers)
    sleep(3)
    print('Response_code ',resp.status_code)
    soup = bs(resp.content,'html.parser')
    return soup
    
```

```python
def get_data(url):
    print('call get_data url')
    soup= get_soup(url)
    print(soup.title.text.strip())
    table = soup.find('table', class_="tableagmark_new")
    headers= []

    for i in table.find_all('th'):
        title = i.text.strip()
        headers.append(title)
        
    list_of_2d_data= []
    for row in table.find_all('tr')[1:]:
        list_of_data= []
    #     print(row)
        for col in row.find_all('td'):
            list_of_data.append(col.text.strip())
        list_of_2d_data.append(list_of_data)
        
    return list_of_2d_data,headers
    
```

```python
def create_empty_data_frame():
    print('Create empty  data frame')
    url= 'https://agmarknet.gov.in/SearchCmmMkt.aspx?Tx_Commodity=24&Tx_State=UP&Tx_District=1&Tx_Market=2551&DateFrom=06-Jul-2021&DateTo=16-Sep-2021&Fr_Date=06-Jul-2021&To_Date=16-Sep-2021&Tx_Trend=0&Tx_CommodityHead=Potato&Tx_StateHead=Uttar+Pradesh&Tx_DistrictHead=Agra&Tx_MarketHead=Achnera'
    data = get_data(url)
    df= pd.DataFrame(columns =data[1])
    return df
```

```python
def create_data_frame(list_of_2d_data, headers):
    df= pd.DataFrame(list_of_2d_data, columns =headers)
    print('Create data frame')
    return df
```

```python
def append_data_frame(df1,df2):
    df2.append(df1, ignore_index = True)
    print('Append data frame')
    return df2
    
```

```python
def run_setup(list_of_url):
#     data_frame_2 = create_empty_data_frame()
    list_of_data_frames= []
    for url in list_of_url:
        list_of_data_from_get_data = get_data(url) 
        data_frame_1=  create_data_frame(list_of_data_from_get_data[0],list_of_data_from_get_data[1])  
#         data_frame_2= append_data_frame(data_frame_1,data_frame_2)
#         data_frame_2.append(data_frame_1,ignore_index = True)
#         print("run setup")
        
        list_of_data_frames.append(data_frame_1)
    
#         sleep(2)
    return list_of_data_frames
    
```
