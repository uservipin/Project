```python
from bs4 import BeautifulSoup as bs 
import pandas as pd 
import requests 
import time 
```

```python
def fetch_data(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
#     url= "https://wwlinkw.sec.gov/Archives/edgar/data/3662/0000950170-98-000413.txt"
    data = requests.get(url,headers)
    time.sleep(3)
    return data
```

```python
def establish_connection(url):
    data=  fetch_data(url)
    data.status_code
    print(data.status_code)
    while(data.status_code != 200):
        data = fetch_data(url)
        print('data not connected to source')
        if(data.status_code == 200):
            print('welcome data connect successfully')
            data = bs(data.content,'html5lib')
            time.sleep(2)
            break
    return data
```

```python
list_of_href = []

for i in range (1,5):
    url = f'https://www.amazon.in/s?k=plum&i=beauty&rh=n%3A1355016031%2Cn%3A1374407031%2Cp_89%3APlum&dc&page={i}&crid=2Q72BO55L0JXV&qid=1642144726&rnid=1355016031&sprefix=plum+%2Caps%2C906&ref=sr_pg_4'
    data = establish_connection(url)
    
    try: 
        d = data.find_all('div',{"class":"sg-col-4-of-12"})
        for da in data.find_all('div',{"class":"sg-col-4-of-12"}):
            da_1  =  da.find('div', {'class':'s-title-instructions-style'})
            da_2 = da_1.find('a',{'class':"a-link-normal"})['href']
            list_of_href.append('https://www.amazon.in'+da_2)
            print(da_2)
    except:
        print("..........................error")
        pass
```
