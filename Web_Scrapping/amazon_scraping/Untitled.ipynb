```python
from bs4 import BeautifulSoup as bs 
import pandas as pd 
import requests 
import time 
```

```python
list_of_href = pd.read_csv('url.csv')
for i in list_of_href['url']:
    print(i)
```

```python

def fetch_data(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    data = requests.get(url,headers)
    time.sleep(5)
    return data



def establish_connection(url):
    data=  fetch_data(url)
    data.status_code
    print(data.status_code)
    while(data.status_code != 200):
        data = fetch_data(url)
        print('data not connected to source')
        if(data.status_code == 200):
            print('welcome data connect successfully')
            data = bs(data.content,'html5lib')
            time.sleep(2)
            break
    return data



```

```python
list_of_names = []

list_of_MRP = []
list_of_current_price = []
list_of_ext = []

list_of_details = []

list_of_href_20= list_of_href['url'][:20]

for url in list_of_href_20:
    
    print("url :",url)
    data = establish_connection(url)
    
    data_1 = data.find('div',{'id':"ppd"})
    data_2 = data_1.find('div', {'id':"centerCol"}) 

    product_name = data_2.find('div', {'id':"titleSection"})
    product_name = product_name.find('span',{'id':"productTitle"}).text.strip(" ")
    list_of_names.append(product_name)


    count = 1 
    price = data_2.find('div',{'id':"apex_desktop"})
    price_2 = price.find('div',{'id':"corePrice_desktop"})
    for p in price.find_all('span',{'class':'a-offscreen'}):
        t = p.find('span',{'class':'a-offscreen'})
        if count == 1 :
            list_of_current_price.append(p.text)

        if count == 2:
            list_of_ext.append(p.text)

        if count == 3 :
            list_of_MRP.append(p.text)
        count +=1
        print(p.text)


    details = data.find('div',{'id':"featurebullets_feature_div"})
    detail = details.find('ul',{'class':"a-unordered-list a-vertical a-spacing-mini"}).text.strip(' ')
    list_of_details.append(detail)
```
