```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import time
import pandas as pd
```

```python
list_of_href = pd.read_csv('url.csv')
for i in list_of_href['url']:
    print(i)
```

```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
list_of_names = []
list_of_MRP = []
list_of_current_price = []
list_of_ext = []
list_of_details = []

count  = 0
for url in list_of_href['url']:
    time.sleep(1)
    driver.get(url)
    
    item_name = driver.find_element_by_xpath('//*[@id="productTitle"]').text
    list_of_names.append(item_name)
    
    
    try :
        mrp = driver.find_element_by_xpath('//*[@id="corePrice_desktop"]/div/table/tbody/tr[1]/td[2]/span[1]/span[2]').text
    except:
        list_of_MRP.append(" No Mrp")
    
    try:
        discounted_price = driver.find_element_by_xpath('//*[@id="corePrice_desktop"]/div/table/tbody/tr[2]/td[2]/span[1]/span[2]').text 
    except:
        list_of_current_price.append("No discounted price ")
        
    try :
        details = driver.find_element_by_xpath('//*[@id="feature-bullets"]').text    
    except:
        
        list_of_details.append('No detail')
        
    list_of_MRP.append(mrp)
    list_of_current_price.append(discounted_price)
    list_of_details.append(details)
    
    count +=1
    
    print("done", count)
```

```python
dict_of_details = {
    'Product_name':list_of_names,
    "MRP":list_of_MRP,
    "current_MRP":list_of_current_price,
    "product_details": list_of_details
}
```

```python
df = pd.DataFrame.from_dict(dict_of_details,orient='index')
```

```python
df = df.transpose()
csv = df.to_excel("data.xlsx")
```
