```python
data
```

```python
list_of_product_name=[]
list_of_product_mrp=[]
list_of_discounted_price= []
list_of_quantity= []
list_of_brand_id= []

```

```python
for list_data in data:
#     print(list_data)
    for d in list_data:
        list_of_product_name.append(d['name'])
        list_of_product_mrp.append(d['mrp'])
        list_of_discounted_price.append(d['price'])
        list_of_quantity.append(d['quantity'])
        list_of_brand_id.append(d['brand_id'])
```
