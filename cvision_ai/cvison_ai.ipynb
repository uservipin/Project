```python
import  pytesseract
import cv2
import os
import pandas as pd 

pytesseract.pytesseract.tesseract_cmd= r'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
Reg_No = "REGN . NO"
Reg_date= "REG. DT"
Mfg_date= "MFG.DT."
Vin_No = "CH.NO"
Engine_No= "ENO"
Name= "NAME"

List_of_reg_no =[]
list_of_reg_date = []
list_of_vin_no=[]
list_of_eng_no=[]
list_of_name = []

path =r"C:/Users/kumar/OneDrive/Programm/Project/cvision_ai/drive-download-20211117T192647Z-001/"
list_dir =os.listdir(path)
local_folder_name = "drive-download-20211117T192647Z-001/"
for i in list_dir:
#     print(i)
    image_path = local_folder_name+i
    print(image_path)
    img = cv2.imread(image_path)
#     cv2.rectangle(img, (500, 500), (500, 500), [255, 0, 0], 2)
#     cv2.imshow("sample", img)
#     cv2.waitKey(0)
    text = pytesseract.image_to_string(img)
    print(text)
    
    
    list_of_lines= text.split("\n")
    list_of_lines

    for lines in list_of_lines:
        if lines.__contains__(Reg_No):
            reg_no = lines
            reg_no = reg_no[12:22]
            List_of_reg_no.append(reg_no)
            print(reg_no)

        if lines.__contains__(Reg_date):
            reg_date= lines
            reg_date =reg_date[9:18]
            list_of_reg_date.append(reg_date)
            print(reg_date)

        if lines.__contains__(Vin_No):
            vin_no= lines
            vin_no = vin_no[8:27]
            list_of_vin_no.append(vin_no)
            print(vin_no)

        if lines.__contains__(Engine_No):
            eng_no= lines
            eng_no = eng_no[7:19]
            list_of_eng_no.append(eng_no)
            print(eng_no)

        if lines.__contains__(Name):
            name= lines
            name = name[6:]
            list_of_name.append(name)
            print(name)

```
